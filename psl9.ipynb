```python
df = df.drop(columns=['other_wicket_type', 'other_player_dismissed', 'penalty'])
```

```python
df.info()
```

```python
death = df[df['ball'] >= 16]

# Runs per franchise in death overs
death_runs = death.groupby('batting_team')['runs_off_bat'].sum()
print(death_runs.sort_values(ascending=False))
```

```python
death_wicket = death.groupby("batting_team")["player_dismissed"].count()
print(death_wicket.sort_values(ascending=False))
```

```python
death_summary = pd.DataFrame({
    'runs': death_runs,
    'wickets': death_wicket
})

# Runs per wicket = batting value
death_summary['runs_per_wicket'] = death_summary['runs'] / death_summary['wickets']
print(death_summary.sort_values('runs_per_wicket', ascending=False))
```

```python
runs =df.groupby("venue")["runs_off_bat"].sum()
match= df.groupby("venue")["match_id"].nunique()  
avgrunpervenue=(runs)/(match)
```
